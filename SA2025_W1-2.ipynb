### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
# Step 1: Calculate the median mpg
median_mpg = df['mpg'].median()

# Step 2: Filter cars with mpg above the median
above_median_mpg = df[df['mpg'] > median_mpg]

# Step 3: Calculate horsepower-to-weight ratio for those cars
above_median_mpg['hp_to_weight'] = above_median_mpg['horsepower'] / above_median_mpg['weight']

# Step 4: Find the maximum ratio value
max_ratio = above_median_mpg['hp_to_weight'].max()

# Step 5: Get car(s) with the highest ratio
best_ratio_cars = above_median_mpg[above_median_mpg['hp_to_weight'] == max_ratio]


print(best_ratio_cars)

```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import matplotlib.pyplot as plt

# Prepare grouped data
grouped = df.groupby(['model_year', 'origin'])['mpg'].mean().unstack()

# Plot
grouped.plot(kind='line', marker='o', figsize=(10, 6))

plt.title('Average MPG over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Set the visual theme
sns.set(style="whitegrid")

# Define the hue order
hue_order = ['japan', 'europe', 'usa']

# Create the scatterplot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    palette='muted',
    sizes=(20, 200),
    hue_order=hue_order,
    alpha=0.7
)

# Titles and labels
plt.title('Horsepower vs Weight Colored by Origin and Sized by MPG', fontsize=14)
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
df = pd.read_csv('/content/Cars.csv')
print(df.columns)
df['power_to_weight'] = df['horsepower'] / df['weight']

multi_year_cars = df.groupby('name').filter(lambda x: x['model_year'].nunique() > 1)

consistency_stats = multi_year_cars.groupby('name').agg(
    mpg_std=('mpg', 'std'),
    count=('mpg', 'count'),
    avg_mpg=('mpg', 'mean')
).reset_index()

consistent_cars = consistency_stats[consistency_stats['mpg_std'] < 1.0]

consistent_cars_sorted = consistent_cars.sort_values(by=['count', 'avg_mpg'], ascending=[False, False])

print(consistent_cars_sorted[['name', 'count', 'avg_mpg']])



```
